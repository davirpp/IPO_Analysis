# Algoritmo para enviar o dataset vindo da B3, formatá-lo do jeito necessário e salvar em .csv

Fonte: [B3](http://www.b3.com.br/pt_br/)

### A célula abaixo serve para pegar os dados da B3 enviando o dataset bruto e formatá-lo como DataFrame com colunas rotuladas

```python
%%time
import pandas as pd
cota = open("C:/Users/davir/Desktop/COTAHIST_A2019.txt",'r')
head = cota.readline()
print(head)
line = cota.readline()
all_data = []
size = [2, 8, 2, 12, 3, 12, 10, 3, 4, 13, 13, 13, 13, 13, 13, 13, 5, 18, 18, 13, 1, 8, 7, 13, 12, 3]
col = ['tipreg', 'date', 'codbdi', 'codneg', 'tpmerc', 'nomres', 'especi', 'prazot', 'modref', 'preabe', 'premax', 'premin', 'premed', 'preult', 'preofc', 'preofv', 'totneg', 'quatot', 'voltot', 'preexe', 'indopc', 'datven', 'fatcot', 'ptoexe', 'codisi', 'dismes']
while line:
    dt = []
    x = 0
    for s in size: 
        dt.append(line[x:x+s]) 
        x+=s 
    all_data.append(dt)
    line = cota.readline()
    if line.startswith('99'): 
        break

df = pd.DataFrame(data=all_data,columns=col)
```

### Célula abaixo serve para formatar os dados de maneira mais clara e fazendo uma limpeza

```python
col = ['tipreg',  'codbdi',  'tpmerc', 'totneg', 'quatot', 'indopc', 'datven', 'fatcot',  'dismes']
for c in col:
    df[c]=pd.to_numeric(df[c], downcast="integer")

col = ['preabe', 'premax', 'premin', 'premed', 'preult', 'preofc', 'preofv','voltot', 'preexe','ptoexe']
for c in col:
    df[c] = df[c].astype(float)/100

col = ['date','datven']
for c in col:
    df[c] = pd.to_datetime(df[c],format='%Y%m%d',errors='coerce')
    
col = ['codneg']
for c in col:
    df[c] = df[c].str.strip()
```

### Célula abaixo serve para filtrar apenas pelos IPO's

```python
def listaFiltro(dataframe, valores):
    return dataframe.loc[dataframe['codneg'].isin(valores)]

list_ipos_2019 = ["PSVM11", "PARC3", "WIZS3",  "AALR3", "MOVI3", "PARD3", "AZUL4", "CRFB3", "IRBR3", "OMGE3", "CAML3", 
                  "BKBR3", "BRDT3", "GNDI3", "HAPV3", "BIDI4", "SBFG3", "CNTO3","NEOE3", "VIVA3", "BMGB4", "CEAB3", 
                  "MTRE3", "LWSA3", "MDNE3", "PRNR3", "ALPK3", "AMBP3", "SOMA3", "DMVF3", "LJQQ3", "LAVV3", "PGMN3",
                  "PETZ3", "PLPL3", "MELK3", "HBSA3", "CURY3", "BOAS3", "SEQL3", "GMAT3", "TFCO4", "CASH3", "ENJU3", 
                  "AERI3", "RRRP3", "AVLL3", "RDOR3", "NGRD3"]
df = listaFiltro(df, list_ipos_2019)
df
```

```python
# CÉLULA PARA TIRAR COLUNAS INDESEJÁVEIS
df = df.drop(columns=["tipreg", "codbdi", "tpmerc", "nomres", "especi", "prazot", "modref", "premax", "premin", "premed", "preofc", 
                      "preofv", "preexe", "indopc", "datven", "fatcot", "ptoexe", "codisi", "dismes", "totneg", "preabe", "quatot"])
df
```

```python
# CÉLULA PARA RESETAR O INDEX
df = df.reset_index(drop=True)
```

```python
# CÉLULA PARA TRANSFORMAR O DATAFRAME EM CSV
df.to_csv("IPO_2019.csv", index = False)
```
