```python
import pandas as pd
```

```python
df = pd.read_csv("./datasets/Cota_ibov.csv")
```

```python
df
```

```python
def tiraponto(texto):    
    return float(texto.replace('.', '').replace(',', '.'))

def porc(texto):   
    return float(texto.replace(',', '.').replace('%', ''))
```

```python
df['Último'] = df.apply(lambda x: tiraponto(x['Último']), axis=1)
```

```python
df['Var%'] = df.apply(lambda x: porc(x['Var%']), axis=1)
```

```python
d = df[['Data', 'Último', 'Var%']].copy()
d
```

```python
d['Data'] = d.apply(lambda x: pd.to_datetime(x['Data'],format='%d.%m.%Y',errors='coerce'), axis=1)
```

```python
d = d.sort_values(by='Data', ascending = True)
```

```python
d = d.reset_index(drop=True)
```

```python
d['Cumulative'] = 0
```

```python
for i in range((d.index.max()) + 1):
    d.loc[i,'Cumulative'] = ((d.loc[i, 'Último'] / d.loc[0, 'Último']) - 1) * 100
```

```python
d.tail(30)
```

```python
# RESETING INDEX
d = d.reset_index(drop=True)

# TRANSFORMING THE DATAFRAME IN CSV 
d.to_csv(f"COTAHIST_IBOV.csv", index = False)
```
